```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as plotly
import matplotlib
```

```python
gdp = pd.read_csv("/content/gdp.csv")
gdp_per_capita = pd.read_csv("/content/gdp_per_capita.csv")
```

```python
gdp = gdp.drop(['Unnamed: 65'] , axis = 1)
gdp_per_capita = gdp_per_capita.drop(['Unnamed: 65'] , axis = 1)
```

```python
data1 = gdp.melt(id_vars=["Country Name", "Code"], var_name="Year", value_name="GDP")
data1['Year'] = pd.to_numeric(data1['Year'])

data2= gdp_per_capita.melt(id_vars=["Country Name", "Code"], var_name="Year", value_name="GDP per capita")
data2['Year'] = pd.to_numeric(data2['Year'])

cleaned_data = pd.merge(data1, data2, on=['Country Name', 'Year', 'Code'], how='outer', )
```

# **Linear Regression**

```python
from sklearn import linear_model

lin_regression = linear_model.LinearRegression()
lin_regression.fit(new_data_sample[['Year']], new_data_sample.GDP)
```

```python
%matplotlib inline

plt.scatter(new_data_sample.Year, new_data_sample.GDP, color= 'red',s= 30)
plt.plot(new_data_sample[['Year']], lin_regression.predict(new_data_sample[['Year']]), color= 'blue')
plt.xlabel("Year", fontsize= 15)
plt.ylabel("GDP", fontsize= 15)
plt.title(f"{input_country} GDP Scatter Plot")
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X = new_data_sample[['Year']]
y = new_data_sample['GDP']


x_train, x_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 20)

calculation_model = LinearRegression()
calculation_model.fit(x_train, y_train)
y_prediction = calculation_model.predict(x_test)
r2_regression = calculation_model.score(x_test, y_test)

mse = mean_squared_error(y_test, y_prediction)
rmse = mean_squared_error(y_test, y_prediction, squared=False)

input_year= int(input("Please enter the year for which you want to predict GDP: "))
predicted_GDP= calculation_model.predict([[input_year]])

print(f"Predicted GDP for {input_country} in {input_year}:", predicted_GDP[0])
print()
print(f"R2 score: {r2_regression}")

print("Mean Squared Error (MSE): {:.2f}".format(mse))
print("Root Mean Squared Error (RMSE): {:.2f}".format(rmse))
print("Training accuracy: {:.2f}".format(calculation_model.score(x_train, y_train)*100)+'%')
print("Testing accuracy: {:.2f}".format(calculation_model.score(x_test, y_test)*100)+'%')
```

```python
#Linear regression after scaling

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
new_data_sample['GDP'] = scaler.fit_transform(new_data_sample[['GDP']])

x = new_data_sample[['Year']]
y = new_data_sample['GDP']

x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.3, random_state= 20)

calculation_model = LinearRegression()
calculation_model.fit(x_train, y_train)
y_prediction= calculation_model.predict(x_test)
r2_regression= calculation_model.score(x_test, y_test)
print(f"R2 score: {r2_regression}")
print("Training accuracy: {:.2f}".format(calculation_model.score(x_train, y_train)*100)+'%')
```

```python
#Decision Tree (After Scaling):

from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
new_data_sample['GDP'] = scaler.fit_transform(new_data_sample[['GDP']])

x =new_data_sample[['Year']]
y =new_data_sample['GDP']



X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

dtr = DecisionTreeRegressor(max_depth=10)
dtr.fit(X_train, y_train)


y_pred = dtr.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

r2_decision_tree= dtr.score(X_test, y_test)

print(("R2 score: {:.2f}".format((r2_decision_tree)*100))+ "%")
print("Mean Squared Error (MSE): {:.5f}".format(mse))
print("Root Mean Squared Error (RMSE): {:.5f}".format(rmse))
print("Training accuracy: {:.2f}".format(dtr.score(X_train, y_train)*100)+'%')
print("Testing accuracy: {:.2f}".format(dtr.score(X_test, y_test)*100)+'%')
```

# **Random Forest Regressor**

```python
from sklearn.ensemble import RandomForestRegressor

x= new_data_sample[['Year']]
y= new_data_sample['GDP']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)

rfr = RandomForestRegressor(n_estimators=100, random_state=10)
rfr.fit(X_train, y_train)

y_pred= rfr.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

r2_random_forest= rfr.score(X_test, y_test)

predicted_GDP= rfr.predict([[input_year]])

print(f"Predicted GDP per capita for {input_country} in {input_year}:", predicted_GDP[0])
print()

print(("R2 score: {:.2f}".format((r2_random_forest)*100))+ "%")
print("Mean Squared Error (MSE): {:.2f}".format(mse))
print("Root Mean Squared Error (RMSE): {:.2f}".format(rmse))
print("Training accuracy: {:.2f}".format(rfr.score(X_train, y_train)*100)+'%')
print("Testing accuracy: {:.2f}".format(rfr.score(X_test, y_test)*100)+'%')


```

```python
# Random Forest (after scaling):

from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
new_data_sample['GDP'] = scaler.fit_transform(new_data_sample[['GDP']])

x =new_data_sample[['Year']]
y =new_data_sample['GDP']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)

rfr = RandomForestRegressor(n_estimators=100, random_state=10)
rfr.fit(X_train, y_train)

y_pred= rfr.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

r2_random_forest= rfr.score(X_test, y_test)

print(("R2 score: {:.2f}".format((r2_random_forest)*100))+ "%")
print("Mean Squared Error (MSE): {:.5f}".format(mse))
print("Root Mean Squared Error (RMSE): {:.5f}".format(rmse))
print("Training accuracy: {:.2f}".format(rfr.score(X_train, y_train)*100)+'%')
print("Testing accuracy: {:.2f}".format(rfr.score(X_test, y_test)*100)+'%')
```
